```python
import pandas as pd
```

```python
# 1. Create a DataFrame for the 201908-citibike-tripdata data. 
nyc_bicketrip_df = pd.read_csv("201908-citibike-tripdata.csv",low_memory=False)
nyc_bicketrip_df.head()
```

```python
# 2. Check the datatypes of your columns. 
nyc_bicketrip_df.info()
```

```python
# 2b. .
nyc_bicketrip_overanhour = nyc_bicketrip_df.loc[nyc_bicketrip_df['tripduration']>3600]
nyc_bicketrip_overanhour.head()
```

```python
# 3. Convert the 'tripduration' column to datetime datatype.
nyc_bicketrip_df['tripduration'] = pd.to_datetime(nyc_bicketrip_df['tripduration'],unit = 's')
nyc_bicketrip_df.head()
```

```python
# 4. Check the datatypes of your columns. 
nyc_bicketrip_df.info()
```

```python
# 5. Export the Dataframe as a new CSV file without the index.
nyc_bicketrip_df.to_csv("201908-citibike-tripdata_modified.csv",index=False)
```
